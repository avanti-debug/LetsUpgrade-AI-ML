___Linear_regression for Apartment price___

```python
Price vs sqft_living
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 


data=pd.read_excel("C:/Users/Dell/Downloads/Linear Regression.xlsx",sheet_name=0)
X = data["price"].values.reshape(-1, 1) 
Y = data["sqft_living"].values.reshape(-1, 1)  
linear_regressor = LinearRegression()  
linear_regressor.fit(X, Y)  
Y_pred = linear_regressor.predict(X) 
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 


data=pd.read_excel("C:/Users/Dell/Downloads/Linear Regression.xlsx",sheet_name=0)
X = data["price"]
Y = data[["sqft_living","bedrooms","bathrooms","floors"]]
linear_regressor = LinearRegression()  
linear_regressor.fit(X, Y)  
Y_pred = linear_regressor.predict(X) 
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.show()
```
